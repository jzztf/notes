### 制作照片墙

- 图片的坐标

图片的像素坐标从左上角开始，向右为x轴，递增；向下为y轴，递增。

- 复制粘贴多个图片
    - 粘贴透明像素
以小图像素宽高作为步长，遍历大图像素坐标，使用`Image`对象的`paste()`方法进行粘贴


```python
from PIL import Image
im = Image.open('cat.jpg')
# 方法
im.new()
im.open() # 创建Image对象
im.copy()
im.paste()
im.crop((left,upper,right,lower)) # 剪裁
im.convert()
im.save()
im.thumbnail() # 缩略图，参数：包含width和height的元组;返回值为None，改变了Image对象im
im.rotate() # 旋转，参数：0-360
im.transpose() # 旋转，参数：im.ROTATE_90/180/270；im.FLIP_LEFT_RIGHT；im.FLIP_TOP_BOTTOM
im.filter() # 增强，参数：ImageFilter.DETAIL

# 属性
im.size
im.width
im.height
im.filename
im.format
im.format_description
```

### 更多实例 

````python
from PIL import Image

catIm = Image.open('cat.jpg')
catIm_cp = catIm.copy()

faceIm = catIm_cp.crop((200,200,400,400))
f_width,f_height = faceIm.size

multi_face = Image.new('RGB',(1600,1600))
m_width,m_height = multi_face.size

for x in range(0,m_width,f_width):
    for y in range(0,m_height,f_height):
        multi_face.paste(faceIm,(x,y))
        
multi_face.save('multi_face.jpg')
````

````python
f = Image.open('multi_face.jpg')
print(f"{f.filename} {f.size} {f.format} {f.format_description}")
f
````

````python
ff = f.resize((400,400))
ff.save('ff.jpg')
````
